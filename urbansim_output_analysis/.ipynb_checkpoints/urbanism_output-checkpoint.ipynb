```python
# set run numbers for 2010 and scenarios
ten = '494'
scens = {'np': '494', 'th': '422', 'pr': '428', 'au': '432'}
```

```python
np_sd10 = pd.read_csv('./output_csv/run' + ten + '_superdistrict_summaries_2010.csv', index_col='superdistrict' )
```

```python
np_sd10.rename(columns={'tothh' : 'tothh'+ten, 'hhpop' : 'hhpop10', 'hhincq1' : 'hhincq110', 'hhincq2' : 'hhincq210', 'hhincq3' : 'hhincq310', 'hhincq4' : 'hhincq410', 'sfdu' : 'sfdu10', 'mfdu' : 'mfdu10', 'totemp' : 'totemp10'}, inplace = True)
```

```python
np_sd40 = pd.read_csv('./output_csv/run494_superdistrict_summaries_2040.csv', index_col='superdistrict')
```

```python
#label each as np or np40 -- can i make this a var easily?
```

```python
np_sd = pd.concat([np_sd10, np_sd40], axis=1)
```

```python
np_sd
```

```python
np_sd['hhch'] = np_sd.tothh - np_sd.tothh10
```

```python
sum_tothh = np_sd["tothh"].sum()
```

```python
np_sd['hhsh'] = np_sd.tothh10 / sum_tothh
```

```python
np_sd['hhsh']
```

```python
np_sd['sfch'] = np_sd.sfdu - np_sd.sfdu10
```

```python
np_sd['mfch'] = np_sd.mfdu - np_sd.mfdu10
```

```python
np_sd['HUch'] = np_sd.mfch + np_sd.sfch
```
