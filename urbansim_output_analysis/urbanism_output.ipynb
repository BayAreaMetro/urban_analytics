```python
# calculate 2010 vars
sd['hu10'] = sd.mfdu10 + sd.sfdu10
sum_hu10 = sd["hu10"].sum()
sd['sh_hu10'] = sd.hu10 / sum_hu10

```

```python
# calculate np vars

# hh
sd['ch_hhnp'] = sd.tothhnp - sd.tothh10
sum_tothhnp = sd["tothhnp"].sum()
sd['sf_hhnp'] = sd.tothh10 / sum_tothhnp

# hu
sd['hunp'] = sd.mfdunp + sd.sfdunp
sd['ch_hunp'] = sd.hunp - sd.hu10
sum_hunp = sd["hunp"].sum()
sd['sh_hunp'] = sd.hunp / sum_hunp
sd['shch_hunp'] = sd.sh_hunp - sd.sh_hu10
sd['vac_hunp'] = sd.tothhnp / sd.hunp

# emp
sd['ch_empnp'] = sd.totempnp - sd.totemp10
sum_totempnp = sd["totempnp"].sum()
sd['sh_empnp'] = sd.totempnp / sum_totempnp

# nrsf
sd['ch_nrsqftnp'] = sd.nrsqftnp - sd.nrsqft10
sum_nrsqftnp = sd["nrsqftnp"].sum()
sd['sh_nrsqftnp'] = sd.nrsqftnp / sum_nrsqftnp
```

```python
sd_out = sd[['hu10', 'sh_hu10', 'hunp', 'sh_hunp', 'shch_hunp', 'vac_hunp']]
```
