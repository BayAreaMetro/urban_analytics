```python
# read and rename 2010 SD
sd10 = pd.read_csv('./output_csv/run' + ten + '_superdistrict_summaries_2010.csv', index_col='superdistrict' )
sd10.rename(columns={'tothh' : 'tothh10', 'hhpop' : 'hhpop10', 'hhincq1' : 'hhincq110', 'hhincq2' : 'hhincq210', 'hhincq3' : 'hhincq310', 'hhincq4' : 'hhincq410', 'sfdu' : 'sfdu10', 'mfdu' : 'mfdu10', 'totemp' : 'totemp10'}, inplace = True)
```
